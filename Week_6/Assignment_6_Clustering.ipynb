### Task 2: Cluster the creditcard users into sub-groups. Perform a statistical analysis of these groups.
* use a Scikit-Learn clustering clustering to group the users 
* optimize over at lest 3 algorithms and theis hyper-parameters
* analyze the key differences between the user groups: use simple statistics 

```python
X = data
X = X.drop(['CUST_ID'], axis=1)
X
```

####AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clustering_Agg = AgglomerativeClustering().fit(X)
```

```python
clustering_Agg.labels_
```

```python
X_AGG = X
```

```python
X_AGG['Group'] = clustering_Agg.labels_
```

```python
X_AGG['Group'].value_counts()
```

####DBSCAN


```python
from sklearn.cluster import DBSCAN
```

```python
clustering_DBSAN = DBSCAN(eps=100, min_samples=25).fit(X)
```

```python
clustering_DBSAN.labels_
```

```python
X_DBSAN = X
```

```python
X_DBSAN['Group'] = clustering_DBSAN.labels_
```

```python
X_DBSAN['Group'].value_counts()
```

#### Meanshift

```python
from sklearn.cluster import MeanShift
```

```python
clusterin_mean = MeanShift().fit(X)
```

```python
clusterin_mean.labels_
```

```python
X_Mean = X
```

```python
X_Mean['Group'] = clusterin_mean.labels_
```

```python
X_Mean['Group'].value_counts()
```
